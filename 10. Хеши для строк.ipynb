**A. Взлом хеш-функции**

def get_hash(string):

    return result

**Пример**
Ввод:

```python
def get_hash(s):
    res = 0
    for i in s:
        res = (res + ord(i)) % 128
    return res


n = int(input())
s = 'dd'
for i in range(n):
    print(s)
    s += 'dd'
```

**C. Основание строки**
Строка S была записана много раз подряд, после чего от получившейся строки взяли префикс и дали вам. Ваша задача определить минимально возможную длину исходной строки S.

**Формат ввода**
В первой и единственной строке входного файла записана строка, которая содержит только латинские буквы, длина строки не превышает 50000 символов.

**Формат вывода**
Выведите ответ на задачу.

**Пример**
Ввод: 
zzz

Вывод:
1

```python
def get_hash(nd, k):
    hash.append(0)
    data.append(1)
    for i in range(0, len(nd)):
        hash.append((hash[-1] * k + nd[i]) % (10 ** 9 + 7))
        data.append((data[-1] * k) % (10 ** 9 + 7))


def check(f1, f2, k):
    from_1 = hash[f1 + k - 1] - hash[f1 - 1] * data[k]
    from_2 = hash[f2 + k - 1] - hash[f2 - 1] * data[k]
    if from_1 % (10 ** 9 + 7) == from_2 % (10 ** 9 + 7):
        return True
    else:
        return False


word = input()
n = len(word)
a = [ord(i) for i in word]
data, hash = [], []
get_hash(a, 28)
for i in range(1, n + 1):
    flag = 1
    for j in range(1, n + 2 - i, i):
        if check(1, j, i) == 0:
            flag = False
            break
    if check(1, i * (n // i) + 1, n % i) == 0:
        flag = False
    if flag:
        print(i)
        break
```

**D. Сравнение подстрок**
Дана строка S, состоящая из строчных латинских букв.

Рассмотрим две её подстроки длины L, первая из которых начинается в индексе A, а вторая — в индексе B.

Научитесь эффективно определять, какая из этих подстрок лексикографически меньше (то есть следует раньше в алфавитном порядке).

**Формат ввода**
Первая строка содержит последовательность S (1 ≤ |S| ≤ 2 ⋅ 105), состоящую из строчных латинских букв.

Вторая строка содержит целое число Q (1 ≤ Q ≤ 2 ⋅ 105) — количество запросов.

Следующие Q строк описывают запросы. Каждая из них содержит целые числа Li, Ai и Bi (1 ≤ Li ≤ |S|, 0 ≤ Ai, Bi ≤ (|S| - Li)) — соответственно длину подстрок и индексы начала каждой из них.

**Формат вывода**
Для каждого из запросов выведите -1, если первая подстрока лексикографически меньше второй, 0, если подстроки равны, или 1, если первая подстрока лексикографически больше второй.

**Пример**
Ввод: 
caeabaeadedcbdcdccec
10
13 4 3
2 12 15
10 1 3
3 8 15
13 5 6
7 2 6
9 8 8
19 0 0
19 0 0
6 7 13

Вывод:
1
-1
1
1
-1
-1
0
0
0
-1


```python
word = input()
n = int(input())
for _ in range(n):
    first_id, from_1, from_2 = map(int, input().split())
    flag = True
    for i in range(first_id):
        if word[from_1 + i] > word[from_2 + i]:
            flag = False
            print(1)
            break
        elif word[from_1 + i] < word[from_2 + i]:
            flag = False
            print(-1)
            break
    if flag:
        print(0)
```

**E. Z-функция**
Дана непустая строка S, длина которой N не превышает 106. Будем считать, что элементы строки нумеруются от 0 до N-1.

Вычислите z-функцию z[i] для всех i от 0 до N-1. z[i] определяется как максимальная длина подстроки, начинающейся с позиции i и совпадающей с префиксом всей строки. z[0] = 0

**Формат ввода**
Одна строка длины N, 0 < N ≤ 106, состоящая из прописных латинских букв.

**Формат вывода**
Выведите N чисел — значения z-функции для каждой позиции, разделённые пробелом.

**Пример**
Ввод: 
abracadabra

Вывод:
0 0 0 1 0 1 0 4 0 0 1 

```python
word = input()
n = len(word)
data = [0] * n
mx, delta = 10 ** 9, 0
right, left = 0, 0
for i in range(n):
    if mx >= 0:
        delta = 0
    if i <= right:
        delta = data[i - left]

        if i + delta - 1 < right:
            data[i] = delta
            mx = 0
            continue
        else:
            delta = right - i + 1

    while i + delta < n and word[delta] == word[i + delta]:
        delta += 1

    data[i] = delta
    right, left = i + delta - 1, i

for i in range(n):
    print(data[i], end=" ")
```

**F. Кубики в зеркале**
Привидение Петя любит играть со своими кубиками. Он любит выкладывать их в ряд и разглядывать свое творение. Недавно друзья решили подшутить над Петей и поставили в его игровой комнате зеркало. Известно, что привидения не отражаются в зеркале, а кубики отражаются. Теперь Петя видит перед собой N цветных кубиков, но не знает, какие из этих кубиков настоящие, а какие — отражение в зеркале. Выясните, сколько кубиков может быть у Пети. Петя видит отражение всех кубиков в зеркале и часть кубиков, которая находится перед ним. Часть кубиков может быть позади Пети, их он не видит.

**Формат ввода**
Первая строка входного файла содержит число N ( 1 ≤ N ≤ 1000000 ) и количество различных цветов, в которые могут быть раскрашены кубики — M ( 1 ≤ M ≤ 1000000 ). Следующая строка содержит N целых чисел от 1 до M — цвета кубиков.

**Формат вывода**
Выведите в выходной файл все такие K, что у Пети может быть K кубиков

**Пример**
Ввод: 
6 2
1 1 2 2 1 1

Вывод:
3 5 6

```python
def get_hash(r, ln, d):
    return (d[r + ln - 1] - d[r - 1] * z[ln]) % (10 ** 9 + 7)


n, m = map(int, input().split())
m += 1
data = list(map(int, input().split()))
data_1, data_2 = [0] * (n + 1), [0] * (n + 1)
z = [1] * (n + 1)
for i in range(1, n + 1):
    data_1[i], data_2[i] = (data_1[i - 1] * m + data[i - 1]) % (10 ** 9 + 7), (data_2[i - 1] * m + data[n - i]) % (10 ** 9 + 7)
    z[i] = (z[i - 1] * m) % (10 ** 9 + 7)
ans = []
for i in range(n // 2 + 1):
    if get_hash(n - i + 1, i, data_2) == get_hash(i + 1, i, data_1):
        ans.append(n - i)
print(*ans[::-1])
```

**G. A-функция от строчки**
Дана строка S, состоящая из N символов. Определим функцию A(i) от первых i символов этой сроки следующим образом:

A(i) = максимально возможному k, что равны следующие строки:

S[1]+S[2]+S[3]+…+S[k]

S[i]+S[i–1]+S[i–2]+…+S[i–k+1]

где S[i] – i-ый символ строки S, а знак + означает, что символы записываются в строчку непосредственно друг за другом.

Напишите программу, которая вычислит значения функции A для заданной строчки для всех возможных значений i от 1 до N.

**Формат ввода**
В первой строке входного файла записано одно число N (1 ≤ N ≤ 200000). Во второй строке записана строка длиной N символов, состоящая только из больших и/или маленьких латинских букв.

**Формат вывода**
В выходной файл выведите N чисел — значения функции A(1), A(2), …, A(N)

**Пример**
Ввод: 
5
aabaa

Вывод:
1 2 0 1 5

```python
def z_function(s):
    n = len(s)
    data = [0] * n
    mx, delta = 10 ** 9, 0
    right, left = 0, 0
    for i in range(n):
        if mx >= 0:
            delta = 0
        if i <= right:
            delta = data[i - left]

            if i + delta - 1 < right:
                data[i] = delta
                mx = 0
                continue
            else:
                delta = right - i + 1

        while i + delta < n and s[delta] == s[i + delta]:
            delta += 1
        data[i] = delta
        right, left = i + delta - 1, i

    return data


n = int(input())
word = input()
word = word + word[::-1]
res = z_function(word)
for i in range(len(res) - 1, len(res) // 2 - 1, -1):
    print(res[i], end=' ')
```

**H. Подпалиндромы**
Строка называется палиндромом, если она читается одинаково как слева направо, так и справа налево. Например, строки abba, ata являются палиндромами.

Вам дана строка. Ее подстрокой называется некоторая непустая последовательность подряд идущих символов. Напишите программу, которая определит, сколько подстрок данной строки является палиндромами.

**Формат ввода**
Вводится одна строка, состоящая из прописных латинских букв. Длина строки не превышает 100000 символов.

**Формат вывода**
Выведите одно число — количество подстрок данной строки, которые являются палиндромами

**Пример**
Ввод: 
aaa

Вывод:
6

```python
word = input()
n = len(word)
radius_odd = [1] * (n * 2 + 1)
word = ' ' + ' '.join([i for i in word]) + ' '
r, l = 0, 0
n = 2 * n + 1
for i in range(1, n):
    L = R = 0
    if i > r:
        L = R = i
    else:
        mi = r + l - i
        L = i - radius_odd[mi] + 1
        R = i + radius_odd[mi] - 1
        if R > r:
            d = R - r
            R -= d
            L += d
    while (L - 1 >= 0) and (R + 1 < n) and word[L - 1] == word[R + 1]:
        L -= 1
        R += 1

    radius_odd[i] = R - i + 1
    if R > r:
        r, l = R, L
count = 0
for i in radius_odd:
    count += i // 2
print(count)
```
